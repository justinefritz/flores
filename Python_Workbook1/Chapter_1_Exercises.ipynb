#### Exercise 1

Given the code below    

    text = 'Business Analytics'

What will the following expressions produce?
    
    text[4]
    text[2:4]
    text[:4]
    text[4:]
    text[-2]
    text[-4:-2]
    text[-2:]
    text[:-2]
    

```python
text = 'Business Analytics'
print(text)
print('4th element in text = ',text[4])
print('2nd to 4th element in text = ',text[2:4])
print('4 element in left in text = ',text[:4])
print('All element except the first 2 element in text = ',text[4:])
print('2nd to the last element in text = ',text[-2])
print('4th to the 3rd to the last element in text = ',text[-4:-2])
print('2 last element in text = ',text[-2:])
print('All element except the last 2 element in text = ',text[:-2])
```
